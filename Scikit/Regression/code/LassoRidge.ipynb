```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
auto_data = pd.read_csv("../data/imports-85.data", sep=r'\s*,\s*', engine='python', names=["symboling", "normalized-losses",
                                                                                          "make", "fuel-type", "aspiration",
                                                                                          "num-of-doors", "body-style", "drive-wheels",
                                                                                          "engine-location", "wheel-base", "length",
                                                                                          "width", "height", "curb-weight", "engine-type",
                                                                                          "num-of-cylinders", "engine-size", "fuel-system",
                                                                                          "bore", "stroke", "compression-ratio", "horsepower", 
                                                                                          "peak-rpm", "city-mpg", "highway-mpg", "price"])
auto_data.head()
```

 We notice that in the data above missing values are represented by **?**, so we need to replace them with **Nan**

```python
import numpy as np

auto_data = auto_data.replace("?", np.nan)

auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.describe(include="all")
```

```python
auto_data["price"].describe()
```

```python
auto_data["price"] = pd.to_numeric(auto_data["price"], errors="coerce")
```

```python
auto_data["price"].describe()
```

```python
auto_data.head()
```

```python
auto_data = auto_data.drop("normalized-losses", axis=1)
auto_data
```

```python
auto_data.describe()
```

```python
auto_data["horsepower"].describe()
```

```python
auto_data["horsepower"] = pd.to_numeric(auto_data["horsepower"], errors="coerce")
auto_data.describe()
```

```python
auto_data["num-of-cylinders"].describe()
```

```python
cylinder_dict = {
    "eight": 8,
    "five": 5,
    "four": 4,
    "six": 6,
    "three": 3,
    "twelve": 12,
    "two": 2
}
auto_data["num-of-cylinders"].replace(cylinder_dict, inplace=True)
auto_data["num-of-cylinders"].describe()
```

```python
auto_data = pd.get_dummies(auto_data, columns=["make",
                                              "fuel-type",
                                              "aspiration",
                                              "num-of-doors",
                                              "body-style",
                                              "drive-wheels",
                                              "engine-location", 
                                              "engine-type",
                                              "fuel-system"])
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data = auto_data.dropna()
auto_data
```

```python
auto_data[auto_data.isnull().any(axis=1)]
```

```python
from sklearn.model_selection import train_test_split

# get you feature data from the auto data
X = auto_data.drop("price", axis=1)

# get your labels
y = auto_data["price"]

# split your data into training and testing 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
linear_model.score(X_train, y_train)
```

```python
linear_model.coef_
```

```python
feature_names = X.columns
coef = pd.Series(linear_model.coef_, feature_names).sort_values()
coef
```

```python
from sklearn.metrics import mean_squared_error
y_pred = linear_model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[15, 6])
plt.plot(y_pred, label="Predicted")
plt.plot(y_test.values, label="Actual")
plt.ylabel("Price")

plt.legend()
plt.show()
```

```python
r_square = linear_model.score(X_test, y_test)
r_square
```

```python
from sklearn.metrics import mean_squared_error

linear_model_mse = mean_squared_error(y_true=y_test, y_pred=y_pred)

linear_model_mse
```

```python
import math

math.sqrt(linear_model_mse)
```

## Using Lasso for the model

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=10)
lasso_model.fit(X_train, y_train)
```

```python
lasso_model.score(X_train, y_train)
```

```python
coef = pd.Series(lasso_model.coef_, feature_names).sort_values()
print(coef)
```

```python
y_pred_lasso = lasso_model.predict(X_test)

plt.figure(figsize=[15, 6])
plt.plot(y_pred_lasso, label="Predicted")
plt.plot(y_test.values, label="Actual")
plt.ylabel("Price")

plt.legend()
plt.show()
```

```python
lasso_model.score(X_test, y_test)
```

```python
lasso_model_mse = mean_squared_error(y_true=y_test, y_pred=y_pred_lasso)
math.sqrt(lasso_model_mse)
```

## Using Ridge Regression for the model

```python
from sklearn.linear_model import Ridge

Rigde_model = Ridge(alpha=10)

Rigde_model.fit(X_train, y_train)
```

```python
Rigde_model.score(X_train, y_train)
```

```python
coef = pd.Series(Rigde_model.coef_, feature_names).sort_values()
coef
```

```python
y_pred_ridge = Rigde_model.predict(X_test)

plt.figure(figsize=[15, 6])
plt.plot(y_pred_ridge, label="Predicted")
plt.plot(y_test.values, label="Actual")
plt.ylabel("Price")

plt.legend()
plt.show()
```

```python
Rigde_model.score(X_test, y_test)
```

```python
Rigde_model_mse = mean_squared_error(y_test, y_pred_ridge)

math.sqrt(Rigde_model_mse)
```
